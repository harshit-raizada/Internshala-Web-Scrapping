```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

Company = []
Profile = []
Stipend = []
Location = []

for i in range(1, 171):
    url = 'https://internshala.com/internships/page-' + str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    all_info = soup.find_all('div', class_ = 'container-fluid individual_internship visibilityTrackerItem')

    for i in all_info:
        Company.append(i.find('div', class_ = 'heading_6 company_name').text.strip())
        Profile.append(i.find('a', class_ = 'view_detail_button').text.strip())
        Stipend.append(i.find('span', class_ = 'stipend').text.strip())
        Location.append(i.find('a', class_ = 'location_link view_detail_button').text.strip())

print(Company)
print('-'*100)
print(Profile)
print('-'*100)
print(Stipend)
print('-'*100)
print(Location)
```

```python
df = pd.DataFrame({'Company': Company, 'Job Role': Profile, 'Stipend': Stipend, 'Location': Location})

df.sample(15)
```

```python
df.to_csv('Internshala.csv', index = False)
```
